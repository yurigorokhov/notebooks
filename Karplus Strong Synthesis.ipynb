```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import IPython
```

```python
def ks(input_arr, delay, output_length, alpha=1):
    output = np.zeros(output_length)
    for i in range(output_length-1):
        output[i] = (input_arr[i] if i < len(input_arr) else 0) + alpha*(output[i-delay] if delay <= i else 0)
    return np.array(output)
```

```python
f = 16000

noise = np.random.randn(200)
out = ks(noise, len(noise), f*2, alpha=.1)
```

```python
plt.stem(out[0:2000])
```

```python
IPython.display.Audio(out, rate=f)
```
